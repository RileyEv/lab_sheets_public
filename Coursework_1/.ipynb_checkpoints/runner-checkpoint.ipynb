```python
# %matplotlib tk

import sys
from functools import partial

import numpy as np
from matplotlib import pyplot as plt

import utilities


def view_data_segments(xs, ys, fs):
    assert len(xs) == len(ys)
    assert len(xs) % 20 == 0
    len_data = len(xs)
    num_segments = len_data // 20
    colour = np.concatenate([[i] * 20 for i in range(num_segments)])

    for i in range(num_segments):
        xs_section = section(xs, i)
        xs_next_section = section(xs, i)
        x_est = np.linspace(np.min(xs_section), np.max(xs_next_section), 100)
        y_est = [fs[i](x) for x in x_est]
        plt.plot(x_est, y_est)

    plt.set_cmap('Dark2')
    plt.scatter(xs, ys, c=colour)
    plt.show()


def least_squares(xs, ys, degree):
    X = np.vander(xs, N=degree+1, increasing=True)
    Y = ys
    
    
    np.savetxt("xTx.csv", X.T.dot(X), delimiter=",")
    
    print(np.linalg.det(X.T.dot(X)))
    
    A = np.linalg.inv(X.T.dot(X)).dot(X.T.dot(Y))
    return A


def sum_squared_error(xs, ys, f):
    ys_hat = f(xs)
    return np.sum(np.square(ys_hat - ys))


def line(A, x):
    y = 0
    for i in range(len(A)):
        y = y + A[i] * np.power(x, i)

    return y


def section(l, i):
    return l[i * 20:i * 20 + 20]


def lsr(xs, ys):
    fs = {}
    errors = {}

    for degree in range(8, 9):
        A = least_squares(xs, ys, degree)
        f = partial(line, A)
        error = sum_squared_error(xs, ys, f)
        fs[degree] = f
        errors[degree] = error

    return fs, errors

# def sse(xs, ys, f):
#     ys_hat = f(xs)
#     return np.sum(np.square(ys - ys_hat))
#
# def ssto(ys):
#     ys_bar = np.mean(ys)
#     return np.sum(np.square())


filename = "train/noise_3.csv"
xs, ys = utilities.load_points_from_file(filename)
print(xs.dtype)
assert len(xs) == len(ys)
assert len(xs) % 20 == 0

len_data = len(xs)
num_segments = len_data // 20

fs = []
degrees = []
errors = []

for i in range(num_segments):
    f, e = lsr(section(xs, i), section(ys, i))
    i = min(e, key=e.get)

    fs.append(f[i])
    degrees.append(i)
    errors.append(e[i])

# print(degrees)
view_data_segments(xs, ys, fs)
```
